```
# Theory Questions - Python Functions and Concepts

# 1. What is the difference between a function and a method in Python?
# Function: A block of code that performs a specific task and is defined using 'def'.
def greet():
    return "Hello!"

# Method: A function that is associated with an object.
text = "hello"
print(text.upper())  # 'upper()' is a method of string object

# 2. Explain the concept of function arguments and parameters in Python.
# Parameters are variables in the function definition; arguments are the values passed.
def add(a, b):  # 'a' and 'b' are parameters
    return a + b

add(3, 5)  # 3 and 5 are arguments

# 3. What are the different ways to define and call a function in Python?
def square(x):
    return x * x

# Call normally
print(square(4))

# Using lambda function
square_lambda = lambda x: x * x
print(square_lambda(4))

# 4. What is the purpose of the 'return' statement in a Python function?
def multiply(a, b):
    return a * b  # returns the result to the caller

# 5. What are iterators in Python and how do they differ from iterables?
# Iterable: Can be looped over (e.g., list, tuple)
# Iterator: Object with next() method and __iter__()
nums = [1, 2, 3]
it = iter(nums)
print(next(it))  # 1

# 6. Explain the concept of generators in Python and how they are defined.
# Generators yield values one at a time using the 'yield' keyword.
def my_gen():
    yield 1
    yield 2

for i in my_gen():
    print(i)

# 7. What are the advantages of using generators over regular functions?
# - Memory efficient
# - Lazy evaluation
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(3):
    print(num)

# 8. What is a lambda function in Python and when is it typically used?
# Anonymous function used for short operations
add = lambda x, y: x + y
print(add(3, 5))

# 9. Explain the purpose and usage of the 'map()' function in Python.
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
print(squared)  # [1, 4, 9, 16]

# 10. What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python?
from functools import reduce

# map(): applies a function to all items
print(list(map(lambda x: x*2, [1, 2, 3])))  # [2, 4, 6]

# filter(): filters elements based on condition
print(list(filter(lambda x: x > 2, [1, 2, 3, 4])))  # [3, 4]

# reduce(): applies a rolling computation
print(reduce(lambda x, y: x + y, [1, 2, 3, 4]))  # 10

# 11. Using pen & paper write the internal mechanism for sum operation using reduce function on this given list: [4, 7, 1, 2, 13]
from functools import reduce

nums = [4, 7, 1, 2, 13]
sum_result = reduce(lambda x, y: x + y, nums)
print("Sum using reduce:", sum_result)

# Internal steps:
# Step 1: 4 + 7 = 11
# Step 2: 11 + 1 = 12
# Step 3: 12 + 2 = 14
# Step 4: 14 + 13 = 27
# Final Output: 27
```
